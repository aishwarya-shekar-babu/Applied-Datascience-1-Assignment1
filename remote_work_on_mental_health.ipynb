```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('Impact_of_Remote_Work_on_Mental_Health.csv',index_col='Employee_ID')
df.head()
```

```python
print(df['Work_Life_Balance_Rating'].unique())
```

```python

def plot_work_life_chart(df,labels):
    
    count = df['Work_Life_Balance_Rating'].value_counts()
    # Create a pie chart
    plt.figure(figsize=(5,5))
    plt.pie(count, labels=labels, autopct='%1.1f%%',wedgeprops={'edgecolor': 'black', 'linewidth': 1.5} ,startangle=140, colors=['lightblue', 'pink','lightgreen','lightsalmon','yellow'])
    plt.title('Work-Life Balance Rating Distribution',fontweight='bold', fontsize=14, color='darkblue', )
    #plt.subplots_adjust(top=1.05)
    plt.subplots_adjust(left=0.2, right=0.9, top=0.90, bottom=0.1)
    plt.axis('equal')  # Equal aspect ratio ensures that pie chart is a circle.


    description = "This pie chart shows the  work-life balance ratings distribution collected from employees. " \
              "The ratings range from Rating 1(Poor) to Rating 5(Excellent), providing view on employee satisfaction."
    
    plt.text(0, -1.5, description, ha='center', fontsize=10, wrap=True)

    # Show the plot
    plt.show()
```

```python
# Create a DataFrame
#df  = pd.read_csv('Impact_of_Remote_Work_on_Mental_Health.csv',index_col='Employee_ID')
# Call the function to plot the pie chart
labels = [
    'Rating 1 (Poor)',
    'Rating 2 (Fair)',
    'Rating 3 (Average)',
    'Rating 4 (Good)',
    'Rating 5 (Excellent)'
]
plot_work_life_chart(df,labels)
```
